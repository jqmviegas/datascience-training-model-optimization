```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline  
```

```python
path_data = '/Users/jqmviegas/src/datascience-training-model-optimization/database/data4tank.xls'
```

```python
df = pd.read_excel(path_data)
```

```python
df_plot = df.copy()
df_plot.iloc[:, :2] = df.iloc[:, :2]*100000
df_plot.plot(figsize=(15, 8))
```

```python
df['dt'] = pd.date_range(start='2019-01-01', periods=2000, freq='10S')
```

```python
df.columns
```

```python
df = df.rename(columns={
    'u1_f1': 'u1',
    'u2_f1': 'u2',
    'y1_f1': 'y1',
    'y2_f1': 'y2',
    'y3_f1': 'y3',
    'y4_f1': 'y4'
})

df.iloc[:, :2] = df.iloc[:, :2]*100000
```

```python
df_raw = df.melt(id_vars=['dt'])
```

```python
df_raw = df_raw.rename(columns={'variable': 'sensor'})
```

```python
df_raw = df_raw.sort_values(['dt', 'sensor']).reset_index(drop=True)
```

```python
df_raw.to_csv('raw_data_4tanks.csv', index=False)
```
