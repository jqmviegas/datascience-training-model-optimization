```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbs
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
%matplotlib inline
```

```python
path_data = '/Users/jqmviegas/src/datascience-training-model-optimization/database/raw_data.csv'
```

```python
df = pd.read_csv(path_data)
```

```python
df.head()
```

```python
df = df.pivot_table(index='dt', values='value', columns='sensor')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.plot(figsize=(15, 8))
```

```python
sbs.pairplot(df)
```

```python
df[['y1', 'y2', 'y3', 'y4']].plot(figsize=(15, 8))
```

```python
for var in ['y1', 'y2', 'y3', 'y4']:
    df[var].plot(figsize=(15, 4))
    plt.title(var)
    plt.show()
```

```python
plot_acf(df['y3'])
plt.show()
```

```python
plot_pacf(df['y2'])
plt.show()
```

# Conclusões

- Sistema de primeira ordem (dependencia no tempo)
- Três momentos diferentes (mudanças de comportamento)
- Não parece ter ruído
- Existe comportamento dinâmico entre diferentes variáveis (pode ser útil usar passado de outras variáveis, etc.)
