```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
inputs = [
    'i1',
    'i2',
    'i3'
]

outputs = [
    'o1'
]
```

```python
dt_index = pd.date_range(start='2019-01-01', end='2019-01-5', freq='S')
```

```python
data_inputs = np.random.rand(len(dt_index), len(inputs))
```

```python
df_data = pd.DataFrame(index=dt_index, columns=inputs, data=data_inputs)
```

```python
for output in outputs:
    df_data[output] = df_data[inputs].T.sum().values*np.random.rand(len(dt_index))
```

```python
df_data.iloc[:60].plot()
plt.show()
```

```python
df_raw_data = pd.melt(df_data.reset_index().rename(columns={'index': 'dt'}), id_vars='dt')
```

```python
df_raw_data = df_raw_data.rename(columns={'variable': 'sensor'})
```

```python
df_raw_data.to_csv('raw_data_rand.csv', index=None)
```
